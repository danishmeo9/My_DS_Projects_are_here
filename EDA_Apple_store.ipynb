# Data Description

- we got main information about this column that ,mostly developer do not provide thier website link, from `1.2` million data 643988 about 52% developers do not provided the websites. so we have to remove this  column

```python
# drop developer website column
df.drop(columns=['Developer_Website'],inplace=True)
print("Developer_Website column dropped successfully!")
```

#### Change the size bytes to MB

```python
# change the name of the column to Size_Mb 
df.rename(columns={'Size_Bytes': 'Size_Mb'}, inplace=True)

# convert bytes to Mb
df['Size_Mb'] = df['Size_Mb']/1000000\

# round the Size_Mb to 2 decimal places
df['Size_Mb'] = df['Size_Mb'].round(2)
```

# Correlation

- Before filling missing values.we have to see the price column and size column s linked with any column or not

```python
# Make a correlation matrix of numeric columns
plt.figure(figsize=(16, 10)) # make figure size  
numeric_cols = ['Current_Version_Score', 'Reviews', 'Average_User_Rating', 'Price', 'Size_Mb'] # make a list of numeric columns
sns.heatmap(df[numeric_cols].corr(), annot=True) # plot the correlation matrix
```

# Filling  Missing values

```python
# Replacing NaN values in numerical columns with the median
df['Price'] = df['Price'].fillna(df['Price'].median())
df['Size_Mb'] = df['Size_Mb'].fillna(df['Size_Mb'].median())

# Replacing NaN values in categorical/string columns
df['Developer_Url'] = df['Developer_Url'].fillna('Unknown')
df['Released'] = df['Released'].fillna(df['Released'].mode()[0])
df['AppStore_Url'] = df['AppStore_Url'].fillna('Unknown')
df['App_Name'] = df['App_Name'].fillna('Unknown')

# Display the result
print(df.isnull().sum())  # Verify no missing values remain
```

# Time Difference b/w app released app updated

```python
# Ensure 'Released' and 'Updated' are in datetime format
df['Released'] = pd.to_datetime(df['Released'], errors='coerce')
df['Updated'] = pd.to_datetime(df['Updated'], errors='coerce')

# Normalize both columns to tz-naive (remove timezone info)
df['Released'] = df['Released'].dt.tz_localize(None)
df['Updated'] = df['Updated'].dt.tz_localize(None)

# Calculate the time difference in days
df['Time_Difference'] = (df['Updated'] - df['Released']).dt.days
# Compute statistics
avg_time_diff = df['Time_Difference'].mean()
min_time_diff = df['Time_Difference'].min()
max_time_diff = df['Time_Difference'].max()
# Print the results
print(f"Average time difference between release and last update: {avg_time_diff:.2f} days")
print(f"Minimum time difference: {min_time_diff} days")
print(f"Maximum time difference: {max_time_diff} days")
```

This project analyzes a comprehensive dataset from the Apple App Store, consisting of `1,230,376 `rows and `21` columns. The analysis begins with importing libraries and loading the dataset. Every app in the dataset has a corresponding App Store URL. Among the most-reviewed apps, Google LLC, Instagram, TikTok, and Spotify stand out as top performers.
